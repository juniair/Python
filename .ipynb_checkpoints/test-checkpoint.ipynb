```python
import urllib2

source = urllib2.urlopen("http://cse.kut.ac.kr/").read()
#print source
#print source # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것
#print ':'.join(t)
from bs4 import BeautifulSoup

soup = BeautifulSoup(source)
pureWords = soup.text.replace("\n","").replace("\t","").replace("  ","")
wordCount = pureWords.split()
#print pureWords
print wordCount
print '\n'
print 'Total pure words is ' + str(len(wordCount)) +' words'
```

```python
deleteString = ''
startTagCount = source.count('<')
endTagCount = source.count('>')
print startTagCount, endTagCount
temp = source
#for
i = 0
while i < len(source):
    #if
    if source[i] == '<':
        startTagCount -= 1       
        #while
        while True:
            deleteString += source[i]
            
            #if
            if source[i] == '>':
                break
            #end if
            i += 1
        #end while
        
        i += 1
    #end if
    temp = temp.strip(deleteString)
        #tf.write('1.\n'+temp+'\n')
    deleteString = ''
#end for
print temp
```
