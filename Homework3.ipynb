####5. 다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터널과 리스트 내포 방식으로 각각 제시하시오.
 - [참고]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)


```python
def addall(list):
    result = 0
    for i in list:
        if i > 0:
            result += i
    return result

print addall([1,2,3,-10,5,6,7,8,9])
```

```python
def addall(list):
    return sum([x for x in list if x > 0],0)

print addall([1,2,3,-10,5,6,7,8,9])

```

 - for ~ in 리터럴에서 list의 원소를 i에 저장하여 i가 양수인 경우 result에 누적하여 리스트에 모든 누적합을 출력하게 하였습니다.
 - 리스트 내포 방식으로 인자로 얻은 리스트를 원소들 중 양수만 어떤 리스트에 원소롤 옮겨 줍니다. 그리고 내장 함수 sum(list, 초기값)을 활용하여 방금 받은 리스트의 누적합을 구합니다.

```python
def addodd(list):
    result = 0
    for i in list:
        if i > 0 and i % 2 == 1:
            result += i
    return result

print addodd([1,2,-3,-11,5,6,7,8,9])
```

```python
def addodd(list):
    return sum([x for x in list if x > 0 and x % 2 == 1],0)

print addodd([1,2,-3,-11,5,6,7,8,9])
```

 - for ~ in 리터럴에서 list의 원소를 i에 저장하여 i가 양수이고 홀수인 경우 result에 누적하여 리스트에 모든 누적합을 출력하게 하였습니다.
 - 리스트 내포 방식으로 인자로 얻은 리스트를 원소들 중 양수이고 홀수인 원소만을 어떤 리스트에 원소롤 옮겨 줍니다. 그리고 내장 함수 sum(list, 초기값)을 활용하여 방금 받은 리스트의 누적합을 구합니다.

####8. 0부터 시작하여 10,001번째의 소수는 얼마인가?

```python
import math
def isPrime(number):
        integerRange=math.floor(math.sqrt(number))
        x=3
        if number==2:
            return True
        
        if number%2 == 0 or number == 1:
            return False
        
        while x <= integerRange:
            if number%x == 0:
                return False
            x+=2
        
        return True

number = 1
primeCount = 0;

while(True):
    
    if isPrime(number):
        primeCount+=1
        
        if primeCount == 10001:
            break
    
    number+=1
    
print "Solution:",number
```

####10. a + b + c = 1000 인 피타고라스 수 a, b, c는 한 가지 뿐입니다. 이 때, a × b × c 는 얼마입니까?

```python
for i in range(1,1000):
    a = i
    t = 1000-a
    c = int( (pow(a,2) + pow(t,2)) / (2*t) )
    b = t - c
    if pow(a,2) + pow(b,2) == pow(c, 2):
        print "solution:", a*b*c
        break
```

 - a + b + c = 1000에서
 - b + C = 1000 - a
 - b = (1000 - a) - c에서 n-a를 t라고 두면
 - b = t - c 가됩니다.
 - 이것을 피타고라스의 정리에 활용을 하면
 - a^2 + b^2 = c^2
 - a^2 + (t-c)^2 = c^2
 - a^2 + t^2 - 2tc + c^2 = c^2
 - a^2 + t^2 =2tc
 - c = (a^2 + t^2)/2t
 - a 값에 의해 b와 c가 결정이 되므로 a를 활용하여 문제를 풉니다.

  - 요구사항 1. 순수 텍스트 안에 존재하는 단어들에 대해 string 모듈을 활용하여 모든 punctuation (구두문자)를 완벽히 제거하시오.  
예: ['world!', ':', '+hello+', '~python$$$', '=', 'world'] ---> ['world', 'hello', 'python', 'world']

  - 요구사항 2. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.  
예: ['컴퓨터공학부!', '---?', '%지향~', '...'] ---> ['컴퓨터공학부', '지향']
  - 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.  
예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}
  - 요구사항 4. 다음 3개의 사이트에 대한 결과를 출력하시오.  
    http://www.naver.com  
    http://www.daum.net  
    http://www.nytimes.com  

```python
import urllib2
import string
from bs4 import BeautifulSoup

def parserURL(source):
    soup = BeautifulSoup(source)
    pureWords = soup.text.replace("//","")

    pureWords = deleteCDATA(pureWords)

    scrapList = string.punctuation
    for i in range(len(scrapList)):
        pureWords = pureWords.replace(scrapList[i],'')

    wordCount = pureWords.split()

    directory = {}
    for word in wordCount:
        if word in directory:
            directory[word] +=1
        else:
            directory[word] = 1

    for key, value in directory.items():
        print key, value
    return;

def deleteCDATA(pureWords):
    count = pureWords.count('<![CDATA[')
    while count != 0:
    
        startTagCDATA = pureWords.find('<![CDATA[')
        endTagCDATA = pureWords.find(']>')
        pureWords = pureWords[0:startTagCDATA] + pureWords[endTagCDATA+1:]
        count -=1
    
    return pureWords

def deleteScrap(pureWords):
    for i in range(pureWords.count('<')+1):
        pureWords = pureWords.replace(pureWords[pureWords.find('<'):pureWords.find('>')+1], '')
    return pureWords
print "http://www.naver.com"
parserURL(urllib2.urlopen("http://www.naver.com/").read())
print
print "http://www.daum.net"
parserURL(urllib2.urlopen("http://www.daum.net").read())
print
print "http://www.nytimes.com"
parserURL(urllib2.urlopen("http://www.nytimes.com").read())
print

#print source # 추후 숙제 답안 제출 때엔 왼쪽 라인은 주석처리할 것



```

  - 제거후 split을 하여 공백을 기준으로 단어를 나누어 리스트화 시켰습니다.
  - for~in 문을 통하여 리스트에 원소를 하나씩 가져오고 그 원소가 사전에 존재하는 key값으로 존재를 하면 그 key값을 가진 값을 1증가 시키고 그렇지 않으면 새로 사전에 등록하여 초기치인 1로 설정하였습니다.
